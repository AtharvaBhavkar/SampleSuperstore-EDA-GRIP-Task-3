**EXPLORATORY DATA ANALYSIS**

```python
sns.pairplot(df)
```

```python
corr = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#Distribution of regions and categoty
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
textprops = {"fontsize":15}
plt.title("Region Distribution")
plt.pie(df['Region'].value_counts(), labels=df['Region'].value_counts().index,autopct='%1.1f%%', textprops=textprops)
plt.subplot(1,2,2)
plt.title("Category Distribution")
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%', textprops=textprops)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,16))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.legend(bbox_to_anchor =(1,0.5))
plt.show()
```

```python
# computing top categories in terms of sales from first 100 observations
top_category_sales=df.groupby('Category').Sales.sum().nlargest(n=300)
# computing top categories in terms of profit from first 100 observations
top_category_profit=df.groupby('Category').Profit.sum().nlargest(n=300)
plt.style.use('seaborn')
top_category_sales.plot(kind='bar',figsize=(10,5),fontsize=14)
top_category_profit.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.title("Top Category Sales VS Profits", fontsize=15)
plt.ylabel("Profits/Total Sales", fontsize=15)
plt.show()
```

```python
# computing top sub-categories in terms of sales from the first 100 observations
top_subcategory_s = df.groupby("Sub-Category").Sales.sum().nlargest(n=100)

# computing top sub-categories in terms of profits from the first 100 observations 
top_subcategory_p = df.groupby("Sub-Category").Profit.sum().nlargest(n=100)

# plotting the see the above data visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind="bar", figsize=(10,5), fontsize=14)
top_subcategory_p.plot(kind="bar", figsize=(10,5), fontsize=14, color='red')
plt.xlabel("Sub-Category", fontsize=15)
plt.ylabel("Total Sales/Profits", fontsize=15)
plt.title("Top Sub-Category Sales VS Profits", fontsize=15)
plt.show()
```

```python
# A more detailed view

plt.figure(figsize=(14,12))
sub_cat_p=df.groupby('Sub-Category')['Profit'].sum().nlargest(n=100)
sub_cat_p.plot.barh()  # h for horizontal representation
```

**Copiers and Phones have the highest profit and profit and Tables have a NEGATIVE profit**

```python
#Visualising discount
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'], data=df)
```

**Visualising the Sales VS Profit in different States**

```python
# computing top states in terms of sales from the first 10 observations
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profits from the first 10 observations 
top_states_p = df.groupby("State").Profit.sum().nlargest(n=10)

# plotting the see the above data visually
plt.style.use('seaborn')
top_states_s.plot(kind="bar", figsize=(10,5), fontsize=14)
top_states_p.plot(kind="bar", figsize=(10,5), fontsize=14, color='red')
plt.xlabel("States", fontsize=15)
plt.ylabel("Total Sales/Profits", fontsize=15)
plt.title("Top 10 states Sales VS Profits", fontsize=15)
plt.show()
```

**Checking the interdependency of sales, profits and discounts**

```python
plt.style.use('seaborn')
df.plot(kind = 'scatter', figsize=(10,5), x="Sales", y="Profit",  c="Discount",s=20, fontsize=16, colormap='viridis')
plt.ylabel("Total Profits", fontsize=16)
plt.title("Interdependancy of Sales, Profits and Discounts", fontsize=16)

plt.show()
```

**The graph clearly shows that if we give more Discount on our products sales increases but the profit decreases.**
